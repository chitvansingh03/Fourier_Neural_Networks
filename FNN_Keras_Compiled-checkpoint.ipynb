```python
import numpy as np
from random import randint


import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation,Dense
from tensorflow.keras.optimizers import Adam , Adagrad , Adadelta
from tensorflow.keras import backend as K
import matplotlib.pyplot as plt

from scipy.special import erf
from scipy import sqrt
import math
from keras import regularizers
from tensorflow.keras import activations
```

```python
# initalising the variables

m = np.sqrt(5)  # std dev of normal distribution of hidden layer weights

N = 4 # no. of nodes
s = (0.5 + (erf(sqrt((2)*m*(math.pi))))/m*4*sqrt(2*(math.pi))) - (erf(m*(math.pi)/sqrt((2)))/m*sqrt(2*(math.pi)))**2  # variance of final output = var of hidden layer output (theorm)
v = s/(N*(0.5 + (erf(sqrt((2)*m*(math.pi))))/m*4*(sqrt(2*(math.pi))))) # std dev of normal dist from op weights are initialised

pi = np.pi
# Creating artificial training Data


d  = 1000 # no. of data points

# initialising initialising condition for initialiser
n =  np.random.randint(98)
w = np.random.randint(98)
print(n , w)

# Creating inpput training data
input_array = (np.random.uniform(-1,1,size = d).reshape(-1,1))

input_arr = pi*input_array # input_arr is th earray to be used in training model
expected_output = 8*np.cos(4*pi*input_array) + np.sin(2*pi*input_array) + np.sin(pi*input_array)  # True output


#Plotting true output vs input array
plt.plot(input_arr , expected_output , '.')

# Defining custom activation function , cos(x)

def custom_activation(x):
    # Define your custom activation function logic
    return K.cos(x)  # Example: Sine activation function

# Initialising hidden layer weights and output layer weights form normal distributions
from tensorflow.keras.initializers import Initializer

hl_weights = tf.keras.initializers.random_normal(mean = 0 ,stddev = m**2 , seed = n)
op_weights =  tf.keras.initializers.random_normal(mean = 0 ,stddev = v**2 , seed = w)

print(hl_weights, op_weights)
```

```python
# model architecture

# We use "Sequential" which is the primary way of defining a linear model in Keras
# Dense basically defines one layer and all of it's components
# units is the number of nodes in the layer
# giving (1,) to input_shape makes keras automatically adjust to the shape of the input array
#kernel_initializer is used for initialising the weights
# kernel_regularizer is used to introduce regularisaation to the weights of the layer which is chosen as L2 norm with its
# coefficient as 0.01
#bias regularizer is so that constant term of the fourier series goes to zero if we want


model = Sequential([
    Dense(units = N, input_shape = (1,) ,activation = custom_activation, kernel_initializer = hl_weights ,
         kernel_regularizer = regularizers.L1(0.001)),
    Dense(units = 1, activation = None , kernel_initializer = op_weights,
          kernel_regularizer = regularizers.L1(0.001), bias_regularizer = regularizers.L1(0.1))
])

# Describing the architecture of the model
model.summary()
```

```python
# Compiling and fitting the model using Keras' model.compile() and model.fit

# We are using optimizer = Adam
# We are using loss as mean squared error which is a good substitute for sum of squared error which is actually just L2 norm
# metric of how well model is permoring is 'mean absolute error'
# model is fitted for 100 iterations. 
# Verbose = 2 gives maximum knowledge for each iteration like time taken for each step,time taken for each epoch, metrics etc
# Shuffle = true shuffles the data once before uusing it, ensuring the bias is possibly reduced

model.compile(optimizer = Adam(learning_rate = 0.01), loss = tf.keras.losses.MeanSquaredError(),
              metrics = tf.keras.metrics.mean_absolute_error )
history = model.fit(x = input_arr , y = expected_output, epochs = 100
          , shuffle = True, verbose= 2)
```

```python
# Printing the final weights and biases as calculated by the model 

print(model.layers[0].get_weights()[0]) # Hidden layer weights 

print(model.layers[0].get_weights()[1]) # Hidden layer biases

print(model.layers[1].get_weights()[0]) # output layer weights

print(model.layers[1].get_weights()[1]) # output layer bias
```
